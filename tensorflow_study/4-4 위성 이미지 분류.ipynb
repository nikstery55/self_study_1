```python
"""
텐서플로우 데이터 셋에 있는 
위성 이미지로 분류 모델을 생성해본다
"""
import tensorflow as tf
import numpy as np
import json
import matplotlib.pylab as plt

# 텐서플로우 데이터셋 활용
import tensorflow_datasets as tfds
```

```python
# EuroSAT 위성 사진 데이터셋 로드
DATA_DIR = "dataset/"

(train_ds, valid_ds).info = tfds.load("eurosat/rgb", split=["train[:80%]", "train[80%:]"],
                                     shuffle_files=True, #파일을 섞어서 학습 잘되게함,
                                     as_supervised=True, #메타정보 불러옴, False이면 (훈련셋, 검증셋)형태의 투플로 반환
                                     with_info = True,
                                     data_dir=DATA_DIR)

print(train_ds)
print(valid_ds)
"""
코랩에서 실행해봄
같은 오류 - https://github.com/lovedlim/tensorflow/issues
책 작성자의 깃에 가보니 
같은 이슈로 질문 올라옴 
여기는 일단 넘어가기
"""
```
